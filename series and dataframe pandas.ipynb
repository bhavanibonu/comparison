```python
import pandas as pd
from functools import reduce
data = {
 'Numbers': [1, 2, 3, 4, 5],
 'Letters': ['A', 'B', 'C', 'D', 'E']
}
df = pd.DataFrame(data)
sq=df['Numbers'].map(lambda x: x**2)
ev=list(filter(lambda x: x % 2 == 0, df['Numbers']))
po = reduce(lambda x, y: x * y, df['Numbers'])
print("Dataframe:\n",df)
print("\nMap for Squaring:\n",sq)
print("\n Filter:\n",ev)
print("\nReduce for product:\n", po)
```

```python
import pandas as pd
a = pd.Series([10, 20, 30, 40, 50])
b = pd.Series([40, 50, 60, 70, 80])
print("Series A:")
print(a)
print("\nSeries B:")
print(b)
non_com = a[~a.isin(b)].tolist() + b[~b.isin(a)].tolist()
print("Items not common to both Series:")
print(non_com)
print("\nSmallest element in Series A:\n", a.min())
print("\nLargest element in Series A:\n",a.max())
print("\nSum of Series B:\n", b.sum())
print("\nAverage of Series A:\n",a.mean())
print("\nMedian of Series B:\n", a.median())
```

```python
import pandas as pd
from functools import reduce
dict = {"size of house":["100 square ft","200 square ft","300 square ft","400 square ft","500 square ft"],
        "price":[1000000,2000000,3000000,4000000,5000000],
        "gender":["male","male","female","male","female"],
        "occupation":["mining","manufacturing","finance","lectural","mining"],
        "age":[35,56,28,44,38],
        "type of house":["Apartment","villa","villa","duplex","Apartment"],
        "Number of rooms in house":[7,9,9,10,7] }
df = pd.DataFrame(dict)
sq=df["type of house"].map
def greater_than(n):
    return n>2000000
rd = reduce(lambda x,y:x*y, df["age"])
lt=list(filter(lambda x:greater_than(x), df["price"]))
print("DataFrame:\n", df)
print("\n map: \n",sq)
print("\n filter:\n",lt)
print("\n list:\n",rd)
```
