```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/bhava/OneDrive/Pictures/Desktop/GPTC/5th SEM/EXCEL/weight-height.csv")
event_a_column = 'Gender'
event_b_column = 'Values'
conditional_probabilities = data.groupby(event_a_column)[event_b_column].mean()
plt.figure(figsize=(8, 6))  # Set the figure size
sns.barplot(x=conditional_probabilities.index, y=conditional_probabilities.values)
plt.xlabel(event_a_column)
plt.ylabel(f'P({event_b_column} | {event_a_column})')
plt.title('Conditional Probability Chart Example')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility if needed
plt.tight_layout()
plt.show()
```
